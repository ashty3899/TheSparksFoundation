```python
#creating 3 new columns

newdf.loc[:, ('New_Cured','New_Deaths','New_Confirmed')]=[0,0,0]
newdf
```

```python
newdf=newdf.sort_values(by=['State','Date'])
newdf
```

```python
NEWCONFLIST=[]
NEWCURELIST=[]
NEWDEATHLIST=[]
p=-1
for c in newdf.iterrows():
    if p==-1:
        NEWCURELIST.append(c[1][4])
        NEWDEATHLIST.append(c[1][5])
        NEWCONFLIST.append(c[1][6])
        
    elif p[1][3]==c[1][3]:
        NEWCURELIST.append(c[1][4]-p[1][4])
        NEWDEATHLIST.append(c[1][5]-p[1][5])
        NEWCONFLIST.append(c[1][6]-p[1][6])
    else:
        NEWCURELIST.append(c[1][4])
        NEWDEATHLIST.append(c[1][5])
        NEWCONFLIST.append(c[1][6])
    p=c 
```

```python
newdf['New_Cured']=NEWCURELIST
newdf['New_Deaths']=NEWDEATHLIST
newdf['New_Confirmed']=NEWCONFLIST
newdf
```

```python
newdf.loc[newdf.New_Confirmed < 0, 'New_Confirmed'] = 0
newdf.loc[newdf.New_Cured < 0, 'New_Cured'] = 0
newdf.loc[newdf.New_Deaths < 0, 'New_Deaths'] = 0
```
